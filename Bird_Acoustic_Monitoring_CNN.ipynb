```python
import tensorflow as tf

data_dir = 'bird_data/train_images'
batch_size = 32
img_height = 180
img_width = 180

# Load training data (80% for training)
train_ds = tf.keras.utils.image_dataset_from_directory(
  data_dir,
  validation_split=0.2,
  subset="training",
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size)

# Load validation data (20% for testing)
val_ds = tf.keras.utils.image_dataset_from_directory(
  data_dir,
  validation_split=0.2,
  subset="validation",
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size)

# Optimize for performance
train_ds = train_ds.cache().prefetch(buffer_size=tf.data.AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size=tf.data.AUTOTUNE)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

# 1. Define the model with Dropout to stop overfitting
model = models.Sequential([
    layers.Rescaling(1./255, input_shape=(img_height, img_width, 3)),

    # Layer 1
    layers.Conv2D(32, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Dropout(0.2), # Prevents memorizing

    # Layer 2
    layers.Conv2D(64, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Dropout(0.3), # Prevents memorizing

    # Layer 3
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.4), # Prevents memorizing

    # Output Layer
    layers.Dense(num_classes)
])

# 2. Compile the model
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

# 3. Add Data Augmentation to help the model generalize
data_augmentation = tf.keras.Sequential([
  layers.RandomFlip("horizontal"),
  layers.RandomRotation(0.1),
])

# Apply augmentation to the training dataset
train_ds = train_ds.map(lambda x, y: (data_augmentation(x, training=True), y))

# 4. Train the model
epochs = 20 # Increased from 10 to give it more time to learn
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=epochs
)

# 5. Save the improved model
model.save('my_bird_model.keras')
```

```python
import pandas as pd
import librosa
import numpy as np

def generate_activity_log(file_path, model, class_names, segment_length=5):
    # 1. Load the long audio
    y, sr = librosa.load(file_path, sr=32000)

    results = []

    # 2. Slice into segments
    for i in range(0, len(y) - (segment_length * sr), segment_length * sr):
        segment = y[i : i + (segment_length * sr)]

        # 3. Process to Spectrogram (Ensure this matches your training logic!)
        mel_spec = librosa.feature.melspectrogram(y=segment, sr=sr, n_mels=128)
        mel_db = librosa.power_to_db(mel_spec, ref=np.max)

        # 4. Prepare for model (Resize to 180x180 and 3 channels)
        import cv2
        img = cv2.resize(mel_db, (180, 180))
        img_3ch = np.stack([img, img, img], axis=-1)
        img_final = np.expand_dims(img_3ch, axis=0)

        # 5. Predict
        raw_pred = model.predict(img_final, verbose=0)
        probs = tf.nn.softmax(raw_pred[0]).numpy()

        # 6. Store top result
        top_idx = np.argmax(probs)
        results.append({
            'timestamp': f"{i/sr:.1f}s",
            'species': class_names[top_idx],
            'confidence': probs[top_idx]
        })

    # 7. Save to CSV
    df = pd.DataFrame(results)
    df.to_csv('final_bird_activity_log.csv', index=False)
    print("Log saved as 'final_bird_activity_log.csv'")
    return df

# Run it!
df = generate_activity_log('XC1148800 - Common Blackbird - Turdus merula.wav', model, class_names)
```
